We are going to study the survival rate of passengers on titanic and what variables affected survival.

Load the dataset in `titanic.xls`. It contains data on all the passengers that travelled on the Titanic.

```python
from IPython.core.display import HTML
HTML(filename='../data/titanic.html')
```

```python
import pandas as pd
import numpy as np

# you would need xlrd - pip install xlrd
t_file = pd.ExcelFile('../data/titanic.xls')
t_df = t_file.parse("titanic", header=0)
t_df.head()
```

### Women and children first?

*** 1. Use the `groupby` method to calculate the proportion of passengers that survived by sex. (1 pt)*** 

```python
# We are calculating the total number of passenger 
# Total
total=t_df['name'].count()
total
# We are checking data by sex
t_df['name'].groupby(t_df['sex']).count()
```

```python
# Calculating proportion of passengers using the total value calculated earlier
# Survived = 1
df_1 = t_df.query('survived == 1')
((df_1['name'].groupby(t_df['sex']).count())/total)*100
```

*** 2. Calculate the same proportion, but by class and sex. (2 pts)*** 

```python
# Calculated the total by class
# Total
total_class=t_df['pclass'].count()
```

```python
# Calculated proportions by class and sex
((t_df.groupby(['pclass','sex'])['name'].count())/total_class)*100
```

```python
# Calculated proportions for survived passengers grouped by class and sex
# Survived = 1
df_1 = t_df.query('survived == 1')
((df_1.groupby(['pclass','sex'])['name'].count())/total_class)*100
```

*** 3. Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, class and sex. (3.5 pts)***

```python
# Creating age groups based on age.
labels = ['children','adolescents','adult','senior']
t_df['agegroup'] = pd.cut(t_df['age'], bins = [0, 14, 20, 65, np.inf], labels=labels, right=False)

# Checking the created age groups
t_df.head()
```

```python
# Calculating the age total count
total_survival=t_df['survived'].count()
```

```python
# Calculating Survival proportion 
df_2 = t_df.query('survived == 1')
((df_2.groupby(['agegroup','pclass','sex'])['name'].count())/total_survival)*100
```
